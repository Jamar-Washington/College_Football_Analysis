```python
# import pandas for the ETL process
import pandas as pd
```

```python
# load in the data using the raw dataset from github
url = 'https://raw.githubusercontent.com/Jamar-Washington/College_Football_Analysis/main/CFBeattendance.csv'
attendance_df = pd.read_csv(url,encoding='cp1252')
attendance_df
```

```python
# Remove 'Current Wins', 'Current Losses', 'New Coach', 'Tailgating', 'TV', and 'Conference' columns as it is not necessary
# Remove 'Year', 'Month', and 'Day' columns as it would be repetitive
attendance_df = attendance_df.drop(columns=['Current Wins', 'Current Losses', 'New Coach', 'Tailgating', 'TV', 'Conference', 'Date'])
attendance_df
```
